```python
import pandas as pd
import numpy as np
import math
from datetime import datetime, timedelta
```

```python
Data = pd.read_csv("c:/python/ExchangeRate@201807-201901.csv",index_col=False)
Data
```

```python
data = Data["現金"]
```

```python
day = []
for i in range(len(data)):
    d = str(Data["資料日期"][i])
    d = datetime.strptime(d ,'%Y%m%d').date()
    day.append(d)
```

```python
day = pd.to_datetime(day)
```

```python
data.index = pd.Index(day)
data = data.iloc[::-1]
```

```python
SAT = pd.date_range(start = '07/02/2018', end = '01/31/2019', freq='W-SAT')
SAT = SAT.drop(pd.to_datetime('2018-12-22')) 
SAT = SAT.drop(pd.to_datetime('2019-01-19')) 
SUN = pd.date_range(start = '07/02/2018', end = '01/31/2019', freq='W-SUN')
```

```python
day = pd.date_range(start = '09/24/2018', end = '09/24/2018', freq='B')
day = day.append(pd.to_datetime(['2018-10-10'])) 
day = day.append(pd.to_datetime(['2018-12-31'])) 
day = day.append(pd.to_datetime(['2019-01-01'])) 
```

```python
s1 = pd.Series([np.nan]*len(SAT) ,index=SAT)
s2 = pd.Series([np.nan]*len(SUN) ,index=SUN)
```

```python
s3 = pd.Series([np.nan]*len(day) ,index = day)
```

```python
data = data.append(s1)
data = data.append(s2)
data = data.append(s3)
```

```python
data.sort_index(inplace=True)
```

```python
for j in range(len(data)):
    null_value = data[data.isnull().values == True] # 缺失值
    if len(null_value) != 0:
        for i in range(len(null_value)):
            if math.isnan(data[null_value.index[i]+timedelta(-1)]) == False:
                if math.isnan(data[null_value.index[i]+timedelta(1)]) == False: #假如前後值有值，則補平均值
                    data[null_value.index[i]] = (data[null_value.index[i]+timedelta(-1)]+data[null_value.index[i]+timedelta(1)])/2
        data = data.fillna(method = 'ffill', limit = 1) #補前一天
        data = data.fillna(method = 'bfill', limit = 1) #補後一天   
    else:
        break
```

```python
data = pd.DataFrame(data, columns=['現金'])
```
