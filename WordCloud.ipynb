```python
from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import jieba
import numpy as np
from collections import Counter
import pandas as pd
```

```python
df = pd.read_csv("c:/python/news_2016-2018.csv")
```

```python
text = df["content"][2]
```

```python
jieba.set_dictionary('c:/python/jieba/dict.txt.big')
```

```python
stopword_set = set()
with open('c:/python/jieba/stopwords.txt','r', encoding='utf-8') as stopwords:
    for stopword in stopwords:
        stopword_set.add(stopword.strip('\n'))
```

```python
terms = [t for t in jieba.cut(text, cut_all=True) if t not in stopword_set]
```

```python
sorted(Counter(terms).items(), key=lambda x:x[1], reverse=True) 
```

```python
mask = np.array(Image.open(r"cloud.jpg"))
```

```python
my_wordcloud = WordCloud(background_color="white",mask = mask, font_path = 'mingliu.ttc',collocations=False, width=2400, height=2400, margin=2)  
my_wordcloud.generate_from_frequencies(Counter(terms))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(my_wordcloud,interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
#顯示用
plt.show()
```
